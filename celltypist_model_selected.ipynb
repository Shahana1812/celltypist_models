```python
# Get an `AnnData` with predicted labels embedded into the cell metadata columns.
adata = predictions.to_adata()
```

```python
adata.obs.majority_voting.replace('AT2', '?', inplace=True)
```

```python
adata.obs.majority_voting
```

```python
# Check if 'tuft' or 'viral' is part of the 'majority_voting' column
if 'tuft' in adata.obs['majority_voting'].values or 'viral' in adata.obs['majority_voting'].values:
    print("Either 'tuft' or 'viral' is present in the 'majority_voting' column.")
else:
    print("Neither 'tuft' nor 'viral' is present in the 'majority_voting' column.")
```

```python
adata.obs['majority_voting'] = adata.obs['majority_voting'].astype('category')

# Filter and replace values of 'majority_voting' column with corresponding values from 'cell_type' column
for idx, row in adata.obs.iterrows():
    if 'viral' in row['cell_type'] or 'tuft' in row['cell_type']:
        # Add the new categories if they don't exist
        for cat in row['cell_type'].split():
            if cat not in adata.obs['majority_voting'].cat.categories:
                adata.obs['majority_voting'] = adata.obs['majority_voting'].cat.add_categories(cat)
        adata.obs.at[idx, 'majority_voting'] = row['cell_type']
```
